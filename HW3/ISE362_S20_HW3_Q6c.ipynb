```python
import numpy as np
import matplotlib.pyplot as plt

R1 = 0.75
R2 = 0.85
C1 = 600
C2 = 1000

Target = 0.9999

n1 = np.arange(0,8)
n2 = np.arange(0,8)

N1, N2 = np.meshgrid(n1, n2)

Rs_array = 1-((1-R1)**N1)*((1-R2)**N2)
Cost = C1*N1+C2*N2
```

```python
Rs_array
```

This Rs_array reads as the following:
Let the coordinate of the top left cell be (0,0), and bottom right cell be (7,7).
For a cell that has the coordiate of (x,y), it contains the Rs a pure parallel system having x type 1 component and y type 2 component. 

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(12,4))
cbar = ax[0].contourf(n1[3:-1], n2[3:-1], Rs_array[3:-1,3:-1])
fig.colorbar(cbar,ax=ax[0])
# fig.tight

cbar = ax[1].contourf(n1, n2, Cost)
fig.colorbar(cbar,ax=ax[1])

plt.show()
```

We plot the Rs and Cost for all combinantion of (# of type 1 component) and (# of type 2 component). It is easy to see that ahomogeneous solution remains optimal for this setting.
